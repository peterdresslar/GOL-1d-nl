## Experimental Data

The files in `./data` contain table output from BehaviorSpace parameter sweeps on `initial-density` through the models (2d-Life (base model), 1d-2d-Life, 1d-Life). Each experiment was run in 2 permutations for the toroidal wrapping settings of On and Off. Our initial analysis takes the raw BehaviorSpace tables and processes them to:

1. Remove the intermediate steps and output resultant dataframes with only final values for each run
2. Group data by run, allowing for easy processing of statistics for individual `initial-density` settings.

The resultant statistics will appear below.

```python
life_on_df = df_and_stats("../data/2d-Life-ON.csv", "initial-density", "2D (Base Model) Wrapping ON")
print("\n\n")
life_off_df = df_and_stats("../data/2d-Life-OFF.csv", "initial-density", "2D (Base Model) Wrapping OFF")
print("\n\n")
life_1d_2d_on_df = df_and_stats("../data/1d-2d-Life-ON.csv", "initial-density", "1D-2D Life Wrapping ON")
print("\n\n")
life_1d_2d_off_df = df_and_stats("../data/1d-2d-Life-OFF.csv", "initial-density", "1D-2D Life Wrapping OFF")
print("\n\n")
life_1d_on_df = df_and_stats("../data/1d-Life-ON.csv", "initial-density", "1D Life Wrapping ON")
print("\n\n")
life_1d_off_df = df_and_stats("../data/1d-Life-OFF.csv", "initial-density", "1D Life Wrapping OFF")
print("\n\n")

```

With these dataframes, we can easily convert to plots:

```python
fig = plot_from_df_finals(life_on_df, life_off_df, life_1d_2d_on_df, life_1d_2d_off_df, life_1d_on_df, life_1d_off_df)
```

As we can see, the plots allow for comparison across the wrapping on and off versions of each of the three models. With this initial analysis, we can likely confirm that there is model conformity between the three models, and that wrapping behavior is generally operational in each.
